```python
# -----------------------------------------------------------
# Dissertation Project: An Empirical Study on the Classification 
# Performance of Deep Learning vs. Gradient Boosting 
# on heterogeneous tabular data
#
# This module provides functions for all model time results
#
# Author: Adam Mabrouk
# Supervisor: Ben Ralph
# Institution: University of Bath
# Created on: 01/01/2024
# Version: 1.0
# -----------------------------------------------------------

# Pandas version: 2.0.3
# Python version: 3.11.5

# Lending club

import matplotlib.pyplot as plt
import numpy as np

features = {
    'NODE': ['Interest Rate (con)', 'Debt To Income (con)', 'Loan To Yearly Income (con)', 
             'Term (cat)', 'Revolving Utility (con)'],
    
    'TabNet': ['Interest Rate (con)', 'Total Accounts (dis)', 'Loan To Yearly Income (con)', 
               'Install To Month Income (con)', 'Debt To Income (con)'],
    
    'XGBoost': ['Interest Rate (con)', 'Purpose (cat)', 'Employment Length (cat)', 
                'Term (cat)', 'Debt To Income (con)']
}


model_percentages = {'NODE': [100, 100, 86.67, 46.67, 46.67], 
                     'TabNet': [73.33, 53.33, 53.33, 53.33, 46.67], 
                     'XGBoost': [100, 100, 100, 86.67, 73.33]
    }

model_colors = {'NODE': "blue", 'TabNet': "green", 'XGBoost': "red"}

max_features_count = max(len(features_list) for features_list in features.values())

index = np.arange(max_features_count)

bar_height = 0.25
space = 0.05

fig, ax = plt.subplots(figsize=(8,7))

for i, model in enumerate(features):
    model_x = index + (bar_height + space)*i
    ax.bar(model_x, model_percentages[model], bar_height, alpha=0.8, color=model_colors[model], label=model)
    for x, y, f in zip(model_x,model_percentages[model], features[model]):
        ax.text(x - 0.05, 5, f,c="white", rotation=90, fontweight='bold')
        ax.text(x - 0.05, y+3, f"{round(y,2)}%", rotation=90)

ax.set_xlabel('First Five Shap Features')
ax.set_ylabel('Frequency of Feature Use %')
ax.set_title('Lending Club Shap Values\n\nFeature Usage Frequency by Model and Type (Mean)')
ax.set_xticks(index + bar_height + space)
ax.set_xticklabels(index+1)
ax.set_ylim(0,119)
ax.legend(loc='upper right', bbox_to_anchor=(1.12, 1))
plt.tight_layout()
plt.savefig('feature_usage_frequency.png')
plt.show()
```
