```python
adult_income = pd.read_csv('income_evaluation_feature_engineered.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

income_counts = adult_income['income'].value_counts()

income_counts.index = ['\n<=50K' if x == 0 else '\n>50K' for x in income_counts.index.unique()]

plt.figure(figsize=(4, 4))  

bar_width = 0.6  

colors = ['#1f77b4', '#ff7f0e']  

new_labels = ['\nDefault (0)\n(n={})'.format(income_counts.get(0, 0)),
              '\nNo Default (1)\n(n={})'.format(income_counts.get(1, 0))]

bars = plt.bar(range(len(income_counts)), income_counts.values, color=colors[:len(income_counts)], width=bar_width)

plt.xticks(range(len(income_counts)), new_labels, rotation=0, ha="center", fontsize=12)

plt.ylabel('Frequency\n', fontsize=12)
plt.title('Adult Income Class Balance', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
