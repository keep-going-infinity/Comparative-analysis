```python
# -----------------------------------------------------------
# Dissertation Project: An Empirical Study on the Classification 
# Performance of Deep Learning vs. Gradient Boosting 
# on heterogeneous tabular data
#
# This module shows the class imbalance for Heloc
#
# Author: Adam Mabrouk
# Supervisor: Ben Ralph
# Institution: University of Bath
# Created on: 01/01/2024
# Version: 1.0
# -----------------------------------------------------------

# Pandas version: 2.0.3
# Python version: 3.11.5

# Heloc

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

heloc = pd.read_csv('heloc_feature_engineered.csv')

RiskPerformance_counts = heloc['RiskPerformance'].value_counts()

plt.figure(figsize=(4, 4))  

bar_width = 0.6

colors = ['#1f77b4', '#ff7f0e']  
colors = colors * (len(RiskPerformance_counts) // len(colors) + 1)

new_labels = ['\nDefault (0)\n(n={})'.format(RiskPerformance_counts.get(0, 0)),
              '\nNo Default (1)\n(n={})'.format(RiskPerformance_counts.get(1, 0))]

bars = plt.bar(range(len(RiskPerformance_counts)), RiskPerformance_counts.values, color=colors[:len(RiskPerformance_counts)], width=bar_width)

plt.xticks(range(len(RiskPerformance_counts)), new_labels, rotation=0, ha="center", fontsize=12)

plt.ylabel('Frequency\n', fontsize=12)
plt.title('HELOC Class Balance', fontsize=14)  
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
